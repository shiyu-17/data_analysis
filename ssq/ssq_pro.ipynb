```python
g = data.groupby('省份')
data_region = g['期号'].count()
data_region
```

```python
data_region.sum()
```

```python
region = data_region.index.tolist()
count = data_region.values.tolist()
```

```python
new = []
for i in region:
    if i == '新疆' : new.append(i+'维吾尔自治区')
    elif i == '西藏' or i == '内蒙古' : new.append(i+'自治区')
    elif i == '广西' : new.append(i+'壮族自治区')
    elif i == '宁夏' : new.append(i+'回族自治区')
    elif i == '重庆' or i == '北京' or i == '天津' or i == '上海': new.append(i+'市')
    else : new.append(i+'省')
m = (
    Map()
    .add('',[list(z) for z in zip(new,count)],maptype='china')
    .set_global_opts(
        title_opts=opts.TitleOpts(title='彩票中奖分布图'),
        visualmap_opts=opts.VisualMapOpts(max_=120)
    )
)
m.render_notebook()
```
